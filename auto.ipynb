```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Automobile_data.csv")
print(df.info())
df.head()
```

```python
df_s1 = df[df['body-style']=='convertible']
df_s1
```

#### Q.1

```python
df_sol1 = pd.DataFrame(df_s1[['price', 'average-mileage']].mean(),columns=['Average value']).reset_index()
df_sol1
```

```python
df_sol1.rename(columns={'index':'variable'},inplace=True)
df_sol1
```

#### Q.2

```python
df_sol2 = df.sort_values(by=['company','price'],ascending=False)
df_sol2
```

#### Q.3

```python
df_sol3 = df[df['company'].str.lower()=='toyota']
df_sol3.head()
```

#### Q.4

```python
df_sol4 = df[df['price'].max() == df ['price']][['company','price']]
df_sol4
```

#### Q.5 

```python
required_items = ['length','wheel-base','horsepower']
```

```python
df_sol5 = pd.DataFrame()

for i in required_items:
     df_temp = pd.DataFrame(df[df[i]==df[i].max()]['company'])
     df_temp['variable']=i
     df_sol5 = pd.concat([df_sol5,df_temp])
df_sol5    
```

#### Q.6

```python
df.groupby(by=['company'],as_index = False)['price'].mean()
```

```python
df_sol6 = df.groupby(by=['company'],as_index = False)['price'].mean()
df_sol6.sort_values(by=['price'],ascending = False,inplace=True)
df_sol6['price'] =df_sol6['price'].round(2)
df_sol6
```

#### Q.7

```python
condition = (df['wheel-base']<100) & (df['length']<175)
```

```python
df['decision'] = np.where((df['wheel-base']<100) & (df['length']<175),"don't buy",'buys')
df
```

#### Q.8

```python
df_sol8 = df[df['decision']=='buys'][['company','price']]
df_sol8 = df_sol8.groupby(by=['company'],as_index=False)['price'].mean()
df_sol8
```
